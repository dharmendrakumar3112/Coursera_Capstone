```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import io
url='https://cocl.us/Geospatial_data'
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
c
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
r = requests.get(url)
#instiantiating the beautifulsoup using lxml parser
soup = BeautifulSoup(r.content, 'lxml')
#extract the table tag from the html response
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))[0]
df.columns = ['Postal Code', 'Borough', 'Neighbourhood']
#df.columns = ['Postalcode', 'Borough', 'Neighbourhood']
#drop the first record as it was the header in the html response
df=df.drop(df.index[0])
#remove the record that are having Borough field as 'Not assigned'
df.drop(df[df.Borough == 'Not assigned'].index, inplace=True)
#use the group by on postalcode and borough and for each element in grooup concatnet the Neighbouhood field with comma seperated
df['Neighbourhood'] = df.groupby(['Postal Code', 'Borough'])['Neighbourhood'].transform(lambda x: ','.join(x))

#now remove the duplicate record
df=df.drop_duplicates()
# if Neighbourhood is not assigned then replace with borough field value of that row
df.Neighbourhood[df.Neighbourhood == 'Not assigned'] = df.Borough
# print the shape ie diemnesion
print(df.shape)
# reset index
df = df.reset_index(drop=True)
#print dataframe
df
```

```python
df_cd = pd.merge(c,df, how='inner',on='Postal Code')
df_cd
```
